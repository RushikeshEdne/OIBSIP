```python
#Modal Building
X = data.drop('Selling_Price', axis=1)
Y = data['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python


# Assuming X is your original DataFrame
cat_columns = ['Fuel_Type', 'Selling_type', 'Transmission', 'Car_Name']

# Use pd.get_dummies() for one-hot encoding
X_encoded = pd.get_dummies(X, columns=cat_columns, prefix=cat_columns)

# The resulting DataFrame X_encoded will have one-hot encoded columns


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, Y_train)
```

```python
y_pred = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
MSE= mean_squared_error(Y_test, y_pred)
RMSE= sqrt(MSE)
print(f'Linear Regression RMSE: {RMSE}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 6))
plt.scatter(Y_test, y_pred, label='Actual')
#plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], '--', color='red', label='Ideal Line (y=x)')
plt.xlabel('Actual Selling Price')
plt.ylabel('Predicted Selling Price')
plt.title('Actual vs. Predicted Selling Price')
plt.legend()

```
