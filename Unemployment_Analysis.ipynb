```python
data = data.rename(columns={data.columns[3]:'EUR' , data.columns[4]:'EE',data.columns[5]:'ELPR'})
data.head()
```

```python
heat_maps=data[['EUR','EE','ELPR']]
heat_maps=heat_maps.corr()

plt.figure(figsize=(5,4))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps,annot=True,cmap='winter');
```

```python
#umemployement rate according to different regions of India
data.columns=["Region" , "Date","Frequency", "EUR", "EE", "ELPR","Area"]

plt.figure(figsize=(5,5))
plt.title("Rate(Unemp)")
sns.histplot(x="EUR",hue="Region",data=data)
plt.tight_layout()
plt.show()
```

```python
#umemployement rate according to different frequency of India
datacolumns=["Region" , "Date","Frequency", "EUR", "EE", "ELPR","Area"]

plt.figure(figsize=(4,4))
plt.title("Unemployment Rate")
sns.histplot(x="EUR",hue="Frequency",data=data)
plt.show()
```

```python
#umemployement rate according to different regions of India
data.columns=["Region" , "Date","Frequency", "EUR", "EE", "ELPR","Area"]

plt.figure(figsize=(5,5))
plt.title("Rate(Unemp)")
sns.histplot(x="EUR",hue="Area",data=data)
plt.show()
```
