```python
# Assuming data is a Pandas DataFrame
X = df.iloc[:, :-1]  # Features (all rows, all columns except the last one)
y = df.iloc[:, -1]   # Target variable (all rows, the last column)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
```

```python
l.fit(X_train,y_train)
```

```python
y_pred=l.predict(X_test)
```

```python
print(y_pred)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 6))
plt.scatter(y_test, y_pred, label='Predicted')
#plt.plot(X_test,y_pred,color='k',label='predicted')
#plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], '--', color='red', label='Ideal Line (y=x)')
plt.xlabel('Actual ')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
plt.legend()

```

```python
plt.subplot(1,2,2)
sns.regplot(x='Radio',y='Sales',marker='.',data=df,line_kws={'color':'k'},scatter_kws={'color':'y'})
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
MSE= mean_squared_error(y_test, y_pred)
RMSE= sqrt(MSE)
print(f'Linear Regression RMSE: {RMSE}')
```
